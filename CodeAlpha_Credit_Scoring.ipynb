```python
# Import libraries

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score


# Create sample dataset

data = {
    'age': [25, 35, 45, 20, 30, 50, 23, 40, 60, 48],
    'income': [30000, 50000, 70000, 20000, 40000, 80000, 35000, 65000, 90000, 72000],
    'loan_amount': [10000, 20000, 30000, 5000, 15000, 40000, 12000, 28000, 50000, 35000],
    'credit_score': [650, 700, 750, 600, 680, 800, 690, 720, 820, 770],
    'approved': [0, 1, 1, 0, 1, 1, 0, 1, 1, 1]
}

df = pd.DataFrame(data)

print("Dataset:")
print(df)


# Split features and target

X = df[['age', 'income', 'loan_amount', 'credit_score']]

y = df['approved']


# Split training and testing

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)


# Train model

model = LogisticRegression()

model.fit(X_train, y_train)

print("Model trained successfully")


# Prediction

y_pred = model.predict(X_test)


# Accuracy

accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)
```

```python
# Interactive prediction

age = int(input("Enter Age: "))
income = int(input("Enter Income: "))
loan = int(input("Enter Loan Amount: "))
credit = int(input("Enter Credit Score: "))

person = [[age, income, loan, credit]]

result = model.predict(person)

if result[0] == 1:
    print("Loan Approved")
else:
    print("Loan Rejected")
```
